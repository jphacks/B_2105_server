```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
# csv読み込み
df = pd.read_csv("drive/My Drive/data_analysis/test1.csv")
```

```python
# 標準化
scaler = StandardScaler()
scaler.fit(np.array(df))
df_std = scaler.transform(np.array(df))
df_std = pd.DataFrame(df_std,columns=df.columns)
```

```python
# 目的変数(Y)
Y = np.array(df_std['honest'])
```

```python
# 説明変数(X)
col_name = ['anger','contempt','disgust','fear','happiness','sadness','surprise']
X = np.array(df_std[col_name])
```

```python
# データの分割(訓練データとテストデータ)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
# モデル構築　
model = LinearRegression()
```

```python
# 学習
model.fit(X_train, Y_train)
```

```python
#予測

#Y_pred = model.predict(X_test)
```

```python
# 回帰係数
coef = pd.DataFrame({"col_name":np.array(col_name),"coefficient":model.coef_}).sort_values(by='coefficient')
```

```python
# 結果
print("【回帰係数】", coef)
print("【切片】:", model.intercept_)
print("【決定係数(訓練)】:", model.score(X_train, Y_train))
print("【決定係数(テスト)】:", model.score(X_test, Y_test))
print("Y_pred")
```
